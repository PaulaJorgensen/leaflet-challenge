```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_to_load = "data/data.csv"

data = pd.read_csv(file_to_load)
data.head()
```

```python
data.columns
```

```python
sm_data=data.loc[:,["number_killed", "number_injured", "full_date",  
                   "day", "month", "year", "city", "state", "location", "urbansuburbanrural", "age", "gender", "education",
                   "relationship_status", "crime_violence_confact_pct", "trauma_crisis_confact_pct", 
                   "mental_health_confact_pct", "motivation_confact_pct", "social_contagion_confact_pct"]]
sm_data
```

```python
victims=sm_data["number_killed"].count()
# victims
states=sm_data["state"].value_counts()
# states
genders=sm_data["gender"].value_counts()
genders
```

```python
group_state=sm_data.groupby(['state'])
state_count=group_state.count().head(10)
state_count
state_avg=group_state.mean()
state_avg
state_percent=state_count/state_avg *100
state_percent.head(15)
```

```python
group_state["number_killed"].sum()
kill_count=group_state["number_killed"].sum()
kill_count.head()
```

```python
summary = pd.DataFrame({"Number of Shootings":states, "Number of Victims Killed":kill_count})
summary.head()

shooting_summary=summary.sort_values("Number of Victims Killed", ascending=False)
shooting_summary.head()
```

```python
top_states=state_count>10
top_states

state_chart = states.plot(kind='bar')
state_chart.set_xlabel("Age Bracket")

plt.tight_layout()
# plt.savefig("state_chart.png")
plt.show(10)
```

```python
gender_group=sm_data.groupby('gender')
count_gender=gender_group['gender'].count()
# count_gender

count_chart = count_gender.plot(kind='bar')
count_chart.set_xlabel("Gender")
count_chart.set_ylabel("Number")

plt.show()
plt.tight_layout()
```

```python
bins =[0, 17, 24, 38, 59, 200]

group_labels = ["0-17", "18-24", "25-39", "40-60", "60 and over"]
```

```python
sm_data["Age Group"] = pd.cut(sm_data["age"], bins, labels=group_labels)
sm_data.head()
```

```python
ages=sm_data["age"].sum()
# ages
```

```python
age= sm_data.groupby("Age Group")
# print(age["Age Group"].count())
age[["age"]].count()
ages=age[["age"]].count()
ages
```

```python
labels = ["0-17", "18-24", "25-39", "40-60", "60+"]
sizes= [7, 40, 61, 57, 6]
colors = ["red", "orange", "lightcoral", "lightblue", "grey"]
explode = (0.5,0,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
       autopct="%1.1f%%", shadow=True, startangle=95)
plt.title=("Age of Shooters %")
plt.savefig("age_pie.png")
plt.show()
```
